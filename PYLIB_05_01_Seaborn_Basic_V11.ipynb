````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
import seaborn as sns
````

````python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 4*np.sin(x+1.0)

plt.figure(figsize=(10,6))   # 그림의 크기
plt.plot(x, y1)
````

````python
plt.plot(x, y1, x,y2, x, y3)   # 3개의 sin 그래프
````

### tip 데이터로 데이터 살펴보고 인사이트 얻기

````python
sns.set_style("whitegrid")
````

````python
tips = sns.load_dataset("tips")  # 인터넷이 켜져 있어야 함.
tips
````

````python
tips.head()  ## 앞의 데이터 조금만 살펴보기
````

### Boxplot을 이용한 시각화 하기

### 요일별 식사금액은 얼마나 될까?

````python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
````

### 요일별 식사 금액, 그런데 흡연자와 비흡연자를 비교해 보자.

````python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
````

### Regression(회귀선)을 그어서 대략적인 예측을 수행해보자.

### 식사금액과 팁의 상관관계

````python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
````

### 그러면 담배 필 때와 안 피는 사람은?

````python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1")
plt.show()
````

````python
fg = sns.load_dataset("flights")
fg.head(5)
fg
````

````python
type(fg)
````

````python
fgp = fg.pivot(index="month", columns="year", values="passengers")
fgp
````

````python
plt.figure(figsize=(10,8))
sns.heatmap(fgp, annot=True, fmt="d")
plt.show()
````
